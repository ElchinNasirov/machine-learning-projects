```python
# Calculus Basics - Derivatives, Partial Derivatives & Gradients

import sympy as sp
import numpy as np

print("Calculus Basics: Derivatives, Partial Derivatives and Gradients\n")

# 1. Symbolic Derivative (Single Variable)
x = sp.symbols('x')
f = x**2 + 3*x + 2
df = sp.diff(f, x)
print("f(x) =", f)
print("f'(x) =", df)
print("Derivative at x=2:", df.subs(x, 2))

# 2. Partial Derivatives (Multivariable)
x, y = sp.symbols('x y')
f_mult = x**2 + y**2 + 3*x*y
print("\nMultivariable function f(x,y) =", f_mult)

partial_x = sp.diff(f_mult, x)
partial_y = sp.diff(f_mult, y)
print("∂f/∂x =", partial_x)
print("∂f/∂y =", partial_y)

# 3. Gradient (Vector of Partial Derivatives)
gradient = sp.Matrix([partial_x, partial_y])
print("Gradient ∇f =", gradient)

# 4. Numerical Gradient using NumPy (for optimization intuition)
def f_num(z):
    x_val, y_val = z
    return x_val**2 + y_val**2 + 3*x_val*y_val

def numerical_gradient(f, z, h=1e-6):
    grad = np.zeros_like(z, dtype=float)
    for i in range(len(z)):
        z_plus = z.copy()
        z_plus[i] += h
        grad[i] = (f(z_plus) - f(z)) / h
    return grad

point = np.array([1.0, 2.0])
print("\nNumerical gradient at point", point, ":", numerical_gradient(f_num, point))
```
