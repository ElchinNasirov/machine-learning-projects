```python
# Probability Basics - Simulations with NumPy

import numpy as np
import matplotlib.pyplot as plt

print("Probability Simulations\n")

# 1. Coin Flip Simulation
n_flips = 10000
coin_flips = np.random.choice(['Heads', 'Tails'], size=n_flips)
heads_count = np.sum(coin_flips == 'Heads')
print(f"After {n_flips} flips: Heads = {heads_count} ({heads_count/n_flips*100:.1f}%)")

# 2. Normal Distribution Simulation
samples = np.random.normal(loc=170, scale=10, size=10000)  # Mean height 170cm, std 10cm
print(f"\nNormal Distribution - Mean: {samples.mean():.2f}, Std: {samples.std():.2f}")

plt.figure(figsize=(10, 6))
plt.hist(samples, bins=50, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(samples.mean(), color='red', linestyle='dashed', linewidth=2, label=f'Mean = {samples.mean():.1f}')
plt.title('Normal Distribution Simulation (Height in cm)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("\nCompleted!")
```
