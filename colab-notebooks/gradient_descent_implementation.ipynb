```python
# Gradient Descent Implementation from Scratch

import numpy as np
import matplotlib.pyplot as plt

print("Gradient Descent Implementation \n")

# Simple quadratic function: f(x) = x² + 3x + 2
def f(x):
    return x**2 + 3*x + 2

def df(x):                      # Derivative: f'(x) = 2x + 3
    return 2*x + 3

# Gradient Descent parameters
x = 10.0                        #Starting point
learning_rate = 0.1             #Step size (eta)
iterations = 50                 #Number of steps

history = [x]                   # Track x values for visualization

print(f"Starting at x = {x}, f(x) = {f(x):.4f}")

for i in range(iterations):
    gradient = df(x)            # Calculate gradient at current point
    x = x - learning_rate * gradient   # Update x (core gradient descent step)
    history.append(x)

    if i % 5 == 0:              #Print every 5 steps
        print(f"Iteration {i:2d}: x = {x:.4f}, f(x) = {f(x):.4f}, gradient = {gradient:.4f}")

print(f"\nFinal result: x = {x:.4f}, f(x) = {f(x):.4f}")

# Visualization
plt.figure(figsize=(10, 6))
x_vals = np.linspace(-15, 15, 400)
plt.plot(x_vals, f(x_vals), label='f(x) = x² + 3x + 2', color='blue')
plt.scatter(history, [f(val) for val in history], color='red', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Optimization')
plt.legend()
plt.grid(True)
plt.show()

print("\n Completed successfully!")
```
