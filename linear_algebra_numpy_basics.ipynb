```python
# Linear Algebra Basics + NumPy Practice
import numpy as np

print("Linear Algebra + NumPy \n")

# 1. Vectors
v = np.array([3, 4])
w = np.array([1, 2])
print("Vector v:", v)
print("Vector w:", w)

# 2. Vector addition
sum_vw = v + w
print("v + w =", sum_vw)

# 3. Dot product
dot = np.dot(v, w)
print("Dot product v·w =", dot)

# 4. Matrix
matrix = np.array([[1, 2], [3, 4]])
print("\nMatrix:\n", matrix)

# 5. Matrix-vector multiplication
transformed = np.dot(matrix, v)
print("Matrix × v =", transformed)

# 6. Determinant
det = np.linalg.det(matrix)
print("Determinant =", det)
```
